```python
with open("/content/ai.txt") as f:
  data=f.read()
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20,
    length_function=len,
    is_separator_regex=False,
)
```

```python
texts = text_splitter.create_documents([data])
print(texts[0])
print(texts[1])
```
