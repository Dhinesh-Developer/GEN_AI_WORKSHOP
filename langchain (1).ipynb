```python
from transformers import AutoTokenizer,AutoModelForTableQuestionAnswering
```

```python
model = 'google/tapas-large-finetuned-wtq'
```

```python
tokenzier = AutoTokenizer.from_pretrained('google/tapas-large-finetuned-wtq')
```

```python
pipelines = pipeline(task="table-question-answering",model="google/tapas-large-finetuned-wtq")
```

```python
model=AutoModelForTableQuestionAnswering.from_pretrained(model)
```

```python
pipelines=pipeline(task="table-question-answering",model=model,tokenizer=tokenzier)
```
